```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

Commentaire

```python
# Chargement des données
job_df = pd.read_csv('Data/jobs.csv')
user_df = pd.read_csv('Data/users.csv')
user_history_df = pd.read_csv('Data/users_history.csv')
feedbacks_df = pd.read_csv('Data/feedbacks.csv')
test_users_df = pd.read_csv('Data/test_users.csv')
```

commentaire

```python
# Fusion des données utilisateur
users_merged_df = pd.merge(user_df, user_history_df, on='UserID', how='left')
```

Commentaire

```python
# Prétraitement des données
job_df['Description'].fillna('', inplace=True)
users_merged_df['JobTitle'].fillna('', inplace=True)
```

Commentaire

```python
# Création de la matrice TF-IDF pour les postes
tfidf = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=0, stop_words='english')
job_tfidf_matrix = tfidf.fit_transform(job_df['Description'])
```

Commentaire

```python

# Fonction de recommandation basée sur le contenu
def recommend_jobs_content_based(user_id, top_n=10):
    user_history = user_history_df[user_history_df['UserID'] == user_id]
    if user_history.empty:
        print(f"No job history found for user {user_id}.")
        return pd.DataFrame(columns=['JobID', 'Title'])
    
    user_job_titles = user_history['JobTitle'].tolist()
    user_tfidf_matrix = tfidf.transform(user_job_titles)
    
    if not user_tfidf_matrix.nnz:
        print(f"No job titles found in history for user {user_id}.")
        return pd.DataFrame(columns=['JobID', 'Title'])
    
    cosine_similarities = linear_kernel(user_tfidf_matrix, job_tfidf_matrix)
    similar_indices = cosine_similarities[0].argsort()[:-top_n-1:-1]
    
    job_ids = [job_df.iloc[i]['JobID'] for i in similar_indices]
    recommended_jobs = job_df.loc[job_df['JobID'].isin(job_ids)]
    recommended_jobs = recommended_jobs[['JobID', 'Title']].reset_index(drop=True)
    
    return recommended_jobs
```

```python
# Fonction de recommandation basée sur la collaboration
def recommend_jobs_collaborative_filtering(user_id, top_n=10):
    user_index = user_df.index[user_df['UserID'] == user_id].tolist()
    if not user_index:
        print(f"User {user_id} not found in user_df.")
        return pd.DataFrame(columns=['JobID', 'Title'])
    
    user_index = user_index[0]
    sorted_user_predictions = predictions_df.iloc[user_index].sort_values(ascending=False)
    sorted_user_predictions = sorted_user_predictions[sorted_user_predictions > 0]
    
    if sorted_user_predictions.empty:
        print(f"No job predictions found for user {user_id}.")
        return pd.DataFrame(columns=['JobID', 'Title'])
    
    job_ids = sorted_user_predictions.index[:top_n].tolist()
    recommended_jobs = job_df.loc[job_df['JobID'].isin(job_ids)]
    recommended_jobs = recommended_jobs[['JobID', 'Title']].reset_index(drop=True)
    
    return recommended_jobs
```

Commentaire

```python
# Boucle de recommandation pour chaque utilisateur de test
for _, user_row in test_users_df.iterrows():
    user_id = user_row['UserID']
    
    content_based_recommendations = recommend_jobs_content_based(user_id)
    collaborative_filtering_recommendations = recommend_jobs_collaborative_filtering(user_id)
    
    print(f"Recommandations basées sur le contenu pour l'utilisateur {user_id}:")
    print(content_based_recommendations)
    
    print(f"Recommandations basées sur la collaboration pour l'utilisateur {user_id}:")
    print(collaborative_filtering_recommendations)
    
    print("------------------------------------------------------")
```
